```python
from scipy.integrate import odeint
import numpy as np
#%matplotlib inline
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#Question 2a no pressure drop 80% conversion
#Define all parameters
T = 680 #K
Fao = 50000 #mol/hr
Pao = 10 #atm
R = 8.314
k=0.054 #mol/gcathatm
kb=0.32
X=0.8
alpha = 0.00 #g-1
Wspan = [0,2000000]
inivalues=[0,1.0] #initial value of X,y

# Set up the differenial equations
def pbr(z,W):  
    [X,y]=z #assignment of dependent variables to convenient variable names
    
    pa = Pao*(1 - X)*y/(1+X) #stoichiometry
    neg_ra = (k*pa)/((1+(kb*pa))**2) #rate law
    
    #ODEs:
    dXdW = neg_ra/Fao  #mole balance
    dydW = -(alpha*(1+X))/(2*y)  #Ergun equation no pressure drop
    return dXdW, dydW

# Setup ODE solver    
W = np.linspace(Wspan[0],Wspan[1],200) # independent variable array, increment 1
solver = odeint(pbr,inivalues,W) # solver has format [X,y]

X=solver[:,0]
y=solver[:,1]
f=(1+X)/y #Calculate f for each element in X and y vectors

# Plot results
plt.plot(W,X, label='X')
plt.plot(W,y, label='y')
plt.plot(W,f, label='f')
plt.xlabel('W (g)')
plt.ylabel('X, y, f')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.ylim(0,3)
plt.title('x,y,f against W(g)')
plt.show()

#From X vs. W, find W required for 80% conversion
W_requirement=np.interp(0.8,X,W)
print('Reactor reaches a conversion of X = 80% no pressure drop at W =',str(W_requirement),'g')
```

```python
#reaction rate graph
pa = Pao*(1 - X)*y/(1+X) #stoichiometry
neg_ra = (k*pa)/((1+(kb*pa))**2) #rate law 
plt.plot(W,neg_ra,'black')
plt.title('Rate of reaction (mol/g cat h)against W(g)')
plt.xlabel('W (g)')
plt.ylabel('Rate (mol/g cat hr)')
plt.grid(True)
plt.show()
```
